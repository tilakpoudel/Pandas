```python
import warnings
import itertools
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# data = sm.datasets.co2.load_pandas()
# y = data.data
df = pd.read_csv('dataset/Upperfinal.csv',usecols=["Date","LTP"])
# df.head()
y=df
print(y)
```

```python
#setting index as date
y['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
y.index = y['Date']

# The 'MS' string groups the data in buckets by start of the month
y = y['LTP'].resample('MS').mean()

# The term bfill means that we use the value before filling in missing values
# y = y.fillna(y.bfill())

print(y)
```

```python
#setting index as date
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df.index = df['Date']

#plot
plt.figure(figsize=(16,8))
plt.plot(df['LTP'], label='LTP Price history')
plt.legend()
```

```python
#setting index as date values
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df.index = df['Date']

#sorting
data = df.sort_index(ascending=True, axis=0)
print(data)

#creating a separate dataset
new_data = pd.DataFrame(index=range(0,len(df)),columns=['Date', 'LTP'])
# print(new_data)

for i in range(0,len(data)):
    new_data['Date'][i] = data['Date'][i]
    new_data['LTP'][i] = data['LTP'][i]
    
    
    
print(new_data)
```
